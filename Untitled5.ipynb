```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, pi
from scipy.ndimage import correlate
from moviepy.editor import VideoClip, ipython_display
from moviepy.video.io.bindings import mplfig_to_npimage
import copy

import os

from tqdm import trange
#from tqdm.notebook import trange
from IPython.display import clear_output, Video
```

```python
class grayScott():
  def __init__(self, Da, Db, f, k, size, noiseConst = (0,0), length = 2.5, lamb = None):
    self.Da = Da
    self.Db = Db
    self.f = f
    self.k = k
    self.size = size
    self.length = length
    self.deltaX = length / size
    self.noiseConstF = noiseConst[0]
    self.noiseConstK = noiseConst[1]
    if lamb == None:
      self.lamb = self.Da
    else:
      self.lamb = lamb

  def seeding(self, seeding):
    Amatrix = np.ones((self.size, self.size), dtype=np.float64)
    Bmatrix = np.zeros((self.size, self.size), dtype=np.float64)

    if seeding == 'paper':

      rng = np.random.default_rng()

      noiseoffseta = rng.uniform(low = 0.99, high = 1.01, size = (20,20))
      noiseoffsetb = rng.uniform(low = 0.99, high = 1.01, size = (20,20))

      minoffset, maxoffset = self.size//2 - 10, self.size//2 + 10

      Amatrix[minoffset:maxoffset, minoffset:maxoffset] = 0.5 * noiseoffseta
      Bmatrix[minoffset:maxoffset, minoffset:maxoffset] = 0.25 * noiseoffsetb

    elif seeding == 'noise':

      rng = np.random.default_rng()
      Bmatrix = rng.uniform(low = 0, high = .5, size = (self.size, self.size))
      Bmatrix[round(self.size*0.45):round(self.size*0.55),
              round(self.size*0.45):round(self.size*0.55)] = 1

    else:

      Bmatrix[round(self.size*0.45):round(self.size*0.55),
              round(self.size*0.45):round(self.size*0.55)] = 1

    return Amatrix, Bmatrix

  def __singleIter(self, Amatrix, Bmatrix, rng, deltaT, noise):

    convolutionMatrix = np.array([[.05, .2, .05],
                                  [.2,-1,.2],
                                  [.05, .2, .05]])

    
    tau = 1 #correlation time.
    sqrt2d = 2 * tau
    Dcsi = self.lamb * self.lamb / (2 * tau)

    noiseComp1 = Dcsi * correlate(noise, convolutionMatrix, mode = 'wrap')  / (np.power(self.deltaX, 2)) - (1/tau) * noise
    noiseComp2 = sqrt2d/tau * rng.normal(size = (self.size, self.size)) * np.sqrt(deltaT)

    noiseNew = np.sin(noise + deltaT * (noiseComp1) + noiseComp2)

    diffusionA = (self.Da * correlate(Amatrix, convolutionMatrix, mode = 'wrap')) / (np.power(self.deltaX, 2))
    diffusionB = (self.Db * correlate(Bmatrix, convolutionMatrix, mode = 'wrap')) / (np.power(self.deltaX, 2))

    abb = Amatrix * Bmatrix * Bmatrix

    fTemp = self.f *(1 + (self.noiseConstF * noiseNew))
    kTemp = self.k *(1 + (self.noiseConstK * noiseNew))

    #fTemp = f
    #kTemp = k

    feed = fTemp * (1 - Amatrix)
    kill = (kTemp + fTemp) * Bmatrix

    Amatrix = Amatrix + (diffusionA - abb + feed) * deltaT
    Bmatrix = Bmatrix + (diffusionB + abb - kill) * deltaT

    return Amatrix, Bmatrix, noiseNew


  def simulate(self, deltaT, period, seeding = None, niter = None, timespan = None, threshold = 1e-8, verbose = True):


    if timespan != None:
      niter = int(timespan//deltaT)


    if verbose:
      if timespan != None:
        print(f"Timespan Given; \n# of Iterations: {niter}")
      elif niter != None:
        print(f"# of Iterations Given; \n# of Iterations: {niter}")
      else:
        niter = 1000

    assert period < niter

    Amatrix, Bmatrix = self.seeding(seeding)

    As = np.zeros((niter//period + 1, self.size, self.size), dtype=np.float64)
    Bs = np.zeros((niter//period + 1, self.size, self.size), dtype=np.float64)

    As[0] = Amatrix.copy()
    Bs[0] = Bmatrix.copy()

    rng = np.random.default_rng(seed = 0)

    counter = 0

    if verbose:
      iters = trange(niter)
    else:
      iters = range(niter)

    noise = rng.uniform(size = (self.size, self.size))

    for i in iters:

      #if i % period == 0:
      #  print(f"Iteration #: {i}")

      AOld = Amatrix.copy()

      Amatrix, Bmatrix, noise = self.__singleIter(Amatrix, Bmatrix, rng, deltaT, noise)

      if np.abs(AOld - Amatrix).sum() <= threshold:
        break


      if i % period == 0:
        As[counter] = Amatrix.copy()
        Bs[counter] = Bmatrix.copy()
        counter += 1

    self.As, self.Bs = As, Bs

    if verbose:
      print("Done!")
    return As, Bs


  def createMovie(self, period, fps = 15, name = 'sim.gif', verbose = True):
    arr = self.As
    duration = arr.shape[0] // fps

    fig, ax = plt.subplots()
    def make_frame(t: int):
        mat = self.As[round(t * fps)]
        ax.clear()
        ax.set_title(f"Da = {self.Da}, Db = {self.Db}, f = {self.f}, k = {self.k}\nNoise = {self.noiseConstF},{self.noiseConstK}")
        ax.margins(x=0)
        ax.set_axis_off()
        ax.imshow(mat, cmap = 'YlOrRd', vmin=0, vmax=1)
        ax.text(20, 20,f"{int(t * fps * period)}")
        return mplfig_to_npimage(fig)

    animation = VideoClip(make_frame, duration=duration)
    animation.write_videofile(f"{name}.mp4",audio=False,
                             fps = fps,
                             verbose=False,
                             logger=None
                             )

    if verbose:
      print("Done!")


```

```python
def multipleSims(parsClass, parsSim, nSims = 100):

  Da, Db, f, k, s = parsClass
  dT, p, timespan = parsSim

  champions = np.empty([5,5], dtype = grayScott)
  meanLengths = np.zeros([5,5])

  
  if not os.path.exists(f"./videos_f{f}_k{k}"):         
      # if the demo_folder directory is not present  
      # then create it. 
      os.makedirs(f"./videos_f{f}_k{k}") 
  
  
  for findex, fnoise in enumerate([0, .25, .50, .75, 1]):
    for kindex, knoise in enumerate([0, .25, .50, .75, 1]):
      noiseConst = (fnoise,knoise)
      print(noiseConst)
      sim = grayScott(Da, Db, f, k, s, noiseConst = noiseConst)
  
      champion = -1
      fullLength = 0
  
      for i in range(nSims):
        print(f"Sim: {i}")
        As, _ = sim.simulate(dT, p,
                    #niter = niter,
                    timespan = timespan,
                    #seeding = 'paper',
                    verbose = False, 
                    )
        
        length = As.shape[0]
        for j in range(As.shape[0])[::-1]:
          if As[j].sum() == 0:
            length -= 1
          else:
            break
  
        print("Length: ",length)
        
        if length > champion:
          champions[findex, kindex] = copy.copy(sim)
          champion = length    
        
        
        fullLength += length
  
      print("Full;:", fullLength)
      meanLengths[findex, kindex] = fullLength / nSims

  
  
      champions[findex, kindex].createMovie(p, 15, f'sim_{fnoise}_{knoise}', verbose = False)
  
  clear_output()
  print(meanLengths)
  os.system('spd-say "your program has finished"')
```

```python
multipleSims(parsClass, parsSim, nSims)
```
