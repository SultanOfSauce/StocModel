```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
nSims = int(1e7)
vals = np.zeros(nSims)

rng = np.random.default_rng()

D = 1
tau = 1
lam = 30

sigma = 2 * np.sqrt(D / (tau * (1 + lam * lam)))

vals = np.sin(2 * pi * rng.normal(size=nSims, scale = sigma))

counts, bins = np.histogram(vals, bins = 100)
plt.stairs(counts, bins)
```
